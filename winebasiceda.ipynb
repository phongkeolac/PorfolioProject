### **Checking correlation**

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='Blues', annot=True)
```

+ Màu tối đại diện cho mối tương quan dương, trong khi đó màu sáng hơn đại diện cho mối tương quan âm

```python
# Quality correlation matrix
k = 12 # number of variables for heatmap
cols = df.corr().nlargest(k, 'quality')['quality'].index
cm = df[cols].corr()
plt.figure(figsize=(12,6))
sns.heatmap(cm, annot=True, cmap = 'viridis')
```

+ Nhìn qua biểu đồ này, ta có thể thấy rằng "density (tỷ trọng)" có mối tương quan dương mạnh với "residual sugar (đường dư)", ngược lại thì nó lại có mối tương quan âm mạnh so với "alcohol"
+ "free sulfur dioxide" và "citric acid" gần như không có mối tương quan với "quality"

### **Checking Outliers**

```python
l = df.columns.values
number_of_columns=12
number_of_rows = len(l)-1/number_of_columns
plt.figure(figsize=(number_of_columns, 5*number_of_rows))
for i in range(0, len(l)):
    plt.subplot(int(number_of_rows + 1), int(number_of_columns), int(i+1))
    sns.set_style('whitegrid')
    sns.boxplot(df[l[i]], color='green')
    plt.tight_layout()
```

+ Ngoại trừ "alcohol" thì tất cả các thuộc tính khác đều có dữ liệu ngoại lai

### **Checking distribution-Skewness**

```python
plt.figure(figsize=(2*number_of_columns,5*number_of_rows))
for i in range(0, len(l)):
    plt.subplot(int(number_of_rows + 1), int(number_of_columns), int(i+1))
    sns.distplot(df[l[i]], kde=True)
```

+ Cột "pH" được phân phối 1 cách bình thường.
+ Biến độc lập ở các cột còn lại đều đều bị sai lệch âm.
